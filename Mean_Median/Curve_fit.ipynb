```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
k=0
z=[4,5,6,7,8]
z=z[k:]
n=[-1.70,-0.9,-1.04,-1.2,-1.56]
n=n[k:]
n_err=[0.65,0.74,0.57,0.16,0.32]
n_err=n_err[k:]
# n=[-1.70,-0.9,-1.04,-4.23,-5.47]
# n_err=[0.65,0.74,0.57,2.12,6.19]
masses=np.arange(7,11.5,0.5)
```

```python
def lin(x,a,b):
    return(a*x+b)
```

```python
val, cov = curve_fit(lin, z,n,sigma=n_err)
```

```python
z_plot=np.arange(0,20)
n_plot=(lin(z_plot,val[0],val[1]))
fig, ax=plt.subplots(figsize=(12,9))
ax.plot(z, n, 'r.', label='Measurements')
ax.errorbar(z, n, n_err, color = 'k', elinewidth = 1, capsize = 2, ls = 'none')
ax.plot(z_plot, n_plot, 'y--', label='Linear fit')
# ax.set(xlabel='Number of independent pointings', ylabel='z',xscale='log',
#        title='Redshift for a mean/median of 1 as a function of pointings')
ax.hlines(n[-1],0,20, linestyle='dashed', label='z=8')
ax.legend()
```
