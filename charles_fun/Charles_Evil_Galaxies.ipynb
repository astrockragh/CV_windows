```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf(M):
    return 10**(-0.6*M)
```

```python
m=np.arange(7,13,0.001)
gal_m=np.random.choice(m, size=int(1e6), p=pdf(m)/sum(pdf(m)))
```

```python
err=np.random.uniform(0.5,1.5,int(1e6))
err=np.random.normal(0,err)
gal_m_err=np.random.choice(m, size=int(1e6), p=pdf(m)/sum(pdf(m)))+err
```

```python
np.mean(err)
```

```python
plt.hist(gal_m)
```

```python
plt.hist(gal_m_err)
```

```python
mask_noerr=gal_m>8.5
mask_err=gal_m_err>8.5
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ye, _, _=ax.hist(gal_m_err[mask_err], range=(8.5, 15), bins=12, label='Observed distribution')
y, _, _=ax.hist(gal_m[mask_noerr], range=(8.5, 15), bins=12, label='True distribution')
ax.set(yscale='log')
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ax.hist(gal_m_err[mask_err], range=(8.5, 15), bins=12, density=1, label='Observed distribution')
ax.hist(gal_m[mask_noerr], range=(8.5, 15), bins=12, density=1, label='True distribution')
ax.set(yscale='log', title='Normalized')
plt.legend()
```
