```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def schechter(z, M_k):
    # Schechter Extrapolation (Bouwens et al. 2015)
    Ms = -20.95 + 0.01 * (z - 6)
    p = 0.47 * 10**(-0.27 * (z - 6)) * 10**(-3)
    a = -1.87 - 0.1 * (z - 6)
    LF_center = np.log(10)/2.5 * p * (10**(0.4 * (Ms - M_k)))**(a+1) * np.exp(-10**(0.4 * (Ms - M_k)))
    return LF_center
```

```python
M=np.linspace(-30,0,1000)
zs=np.arange(20)
fig, ax=plt.subplots(figsize=(20,15))
for z in zs:
    ax.plot(M,schechter(z,M), label=z)
    ax.set(yscale='log')
    ax.legend()
```
