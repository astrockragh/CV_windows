```python
import numpy as np
from astropy.io import fits
from scipy.stats import gamma
import pandas
import matplotlib.pyplot as plt
import sys
%matplotlib inline
plt.rcParams.update({'font.size': 20})
plt.rc('font', family='serif')
plt.rc('text', usetex=True)
sys.path.append('../../../2YR-UNI/Blok2/AppStat2019/External_Functions')
from ExternalFunctions import nice_string_output, add_text_to_ax    
```

```python
def trial(var, mean, trials=10000):
    k = mean**2/var
    t = var/mean
    rand = np.rint(np.random.gamma(k, scale=t, size=trials))
    return rand
```

```python
x=np.linspace(0,20,1000)
mean=np.arange(0,12,3)+1
var=np.arange(4)+1
t=var/mean
k=mean**2/var
fig, ax = plt.subplots(nrows=4,ncols=4, figsize=(15,12))
for i in range(4):
    for j in range(4):
        g=trial(var[j],mean[i],trials=int(1e6))
        ax[i][j].hist(g,bins=np.arange(20),density=1,histtype='step')
        d = {'Made mean:': mean[i],
             'Real Mean:': np.mean(g),
             r'$\sigma_x$ set:': np.sqrt(var[j]),
            r'$\sigma_x$ real:': np.std(g),
            }        
        text = nice_string_output(d, extra_spacing=2, decimals=3)
        add_text_to_ax(0.4, 0.6, text, ax[i][j], fontsize=12)
        
```

```python
def gamma_pmf(x,mean,var):
    k = mean**2/var
    t = var/mean
    rv=gamma(k, scale=t)
    pmf=np.zeros(len(x)-1)
    for i in range(len(x)-1):
        pmf[i]=rv.cdf(x[i+1])-rv.cdf(x[i])
    return pmf
```

```python
fig, ax = plt.subplots(figsize=(15,12))
m, v = 2, 2
x=np.arange(20)
g=trial(v,m,trials=int(1e6))
ax.hist(g,bins=x,density=1,histtype='step')
#         ax[i][j].legend().plotax,
ax.plot(x[1:],gamma_pmf(x,m,v))
d = {'Made mean:': m,
    'Real Mean:': np.mean(g),
    r'$\sigma_x$ set:': np.sqrt(v),
    r'$\sigma_x$ real:': np.std(g),
    }        
text = nice_string_output(d, extra_spacing=2, decimals=3)
add_text_to_ax(0.4, 0.6, text, ax, fontsize=12)
```
